```python
def get_url(begin_date,end_date,api_key):
    NYT_news=[]
    page=0
    for i in range(100):
        page1=str(page)
        print(page1)
        url="http://api.nytimes.com/svc/search/v2/articlesearch.json?fq=source:('The New York Times')&begin_date="+begin_date+"&end_date="+end_date+"&page="+page1+"&type=article&api-key="+api_key   
        print(url)
        r = requests.get(url)
        print(r)
        sample=r.json()
        NYT_news.append(sample)
        page=page+1
        sleep(1)
    return NYT_news    
     
    
```

```python
def make_repo(begin_date,end_date,api_key):
    NYT_news=get_url(begin_date,end_date,api_key)

    k=[]

    for i in range(len(NYT_news)):
        for j in range(len(NYT_news[i]['response']['docs'])):
            p=NYT_news[i]['response']['docs'][j]
            k.append(p)

    z=pd.DataFrame(k)
    
    local_repo=pd.DataFrame({'headline': z.iloc[0]['headline']['main'], 'date': z.iloc[0]['pub_date'][0:10],'url': z.iloc[0]['web_url'],
                             'page': z.iloc[0]['print_page'],'count': z.iloc[0]['word_count'],'section': z.iloc[0]['new_desk'],'id':z.iloc[0]['_id']},index=[0])
    keywords=pd.DataFrame({'id': z.iloc[0]['_id'][0], 'key':z.iloc[0]['keywords'][0]['value'],'rank':z.iloc[0]['keywords'][0]['rank']},index=[0])
    
    for i in range(1,len(z)):
        temp_repo=pd.DataFrame({'headline': z.iloc[i]['headline']['main'], 'date': z.iloc[i]['pub_date'][0:10],'url': z.iloc[i]['web_url'],'page': z.iloc[i]['print_page'],'count': z.iloc[i]['word_count'],'section': z.iloc[i]['new_desk'],'id':z.iloc[i]['_id']},index=[0])
        local_repo = pd.concat([local_repo,temp_repo],ignore_index=True)
        for j in range(1,len(z.iloc[i]['keywords'])):
            temp=pd.DataFrame({'id': z.iloc[i]['_id'], 'key':z.iloc[i]['keywords'][j]['value'],'rank':z.iloc[i]['keywords'][j]['rank']},index=[0])
            keywords = pd.concat([keywords,temp],ignore_index=True)
        
    return local_repo,keywords
    
```

```python
import _pickle as pic
def scrapper(begin_date,end_date,api_key,signature):
    head=[]
    desc=[]
    q=datetime.datetime.strptime(str(begin_date), "%Y%m%d").date()
    month=q.strftime('%B')
    year=q.strftime('%Y')
    local_repo,keywords=make_repo(begin_date,end_date,api_key)
    file_n_l=signature+"_"+month+"_"+year+"_localrepo.csv"
    print(file_n_l)
    local_repo.to_csv(file_n_l, sep=',',encoding='utf-8')
    file_n_k=signature+"_"+month+"_"+year+"_keywords.csv"
    print(file_n_k)
    keywords.to_csv(file_n_k, sep=',',encoding='utf-8')
   
    #print(local_repo)
    for i in range(len(local_repo['url'])):
        print("sequence:"+str(i))
        print(local_repo['url'][i])
        head.append(local_repo['headline'][i])
        req= urllib.request.Request(local_repo['url'][i])
        with urllib.request.urlopen(req) as response:
            the_page = response.read()
        soup = BeautifulSoup(the_page,'html.parser')
        para=[p.text for p in soup.findAll('p',{"class":["story-body-text"]})]
        sent_str = ""
        for j in para:
            sent_str += j
        removeSpecial_para0= sent_str.translate ({ord(c): "" for c in "â€”!@#$%^&*()[]{};:,/'<>?\|`~-=_+"})
        splitted = removeSpecial_para0.split()
        desc.append(splitted[:50])

    file_n=signature+"_"+month+"_"+year+".p"
    print(file_n)
    aFile=open(file_n, 'wb')
    aFile.write(pic.dumps({"head":head,"desc":desc}))
    aFile.close()
    return local_repo,keywords
```

```python
def main():
    begin_date = input("Please the begin_date:")
    end_date = input("Please the end_date:")
    api_key = input("Please enter your api-key:")
    signature = input("Please enter your signature(First-Letter of Name):")
    local_repo,keywords=scrapper(begin_date,end_date,api_key,signature)
```
